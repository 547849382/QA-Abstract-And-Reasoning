### 删掉汉字，英文，数字

```python
# 只保留标点符号
r_only_punctuation = re.compile("[\u4e00-\u9fa5_a-zA-Z0-9]")
# 只保留，！？。.-和汉字英文数字
r_del_punctuation = re.compile("[^，！？。\.\-\u4e00-\u9fa5_a-zA-Z0-9]")
```

```python
# 测试能否删去带括号的 进口 海外
sentence = "这款大众（进口）车(海外)，是从美国进口的海外车"
r = re.compile(r"[(（]进口[)）]|\(海外\)")
sentence = r.sub("", sentence)
sentence
```

```python
# 测试标点符号的转换
# 保留，！？。.-
sentence = "我跟你将，这种**&$@#东西！就是,.哪个。？you-dian123!离谱 dあｓｄｆ"
sentence = sentence.replace(",", "，")
sentence = sentence.replace("!", "！")
sentence = sentence.replace("?", "？")
sentence
```

```python
# 删除其他标点符号
sentence = r_del_punctuation.sub("", sentence)
sentence
```
